```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# 붓꽃 데이터 읽기
iris_data = pd.read_csv('iris.csv', encoding='utf-8')
iris_data
```

# 붓꽃 데이터를 레이블(y, 정답데이터)과 입력데이터(x, 학습데이터)로 분류

```python
# 정답(레이블)
y_label = iris_data.loc[:, 'Name']
y_label
```

```python
# 학습(입력)
x_data = iris_data.loc[:, ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
x_data
```

```python
# 학습전용(Train)과 테스트전용(Test) 분리
x_train, x_test, y_train, y_test = train_test_split(x_data, y_label, test_size=0.2, train_size=0.8, shuffle=True) # shuffle로 학습용과 테스트용 데이터가 겹쳐지지 않도록 섞어준다.
```

```python
x_train
```

```python
# scv 학습
clf = SVC()
clf.fit(x_train, y_train)
```

```python
# 학습된 데이터를 테스트 데이터를 보고 예측
y_predict = clf.predict(x_test)
y_predict
```

```python
# 예측된 데이터를 테스트 데이터의 정답 레이블과 비교하여 정확도를 나타냄
accuracy_score(y_test, y_predict)
```
