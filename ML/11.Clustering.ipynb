```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
```

```python
mglearn.plots.plot_kmeans_algorithm()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

x, y = make_blobs(random_state=1)
# print(x.shape)
print(x[0] ,'->', y[0] )
print(x[10] ,'->', y[10] )
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

```python
print(' : \n{}'.format(kmeans.labels_))
```

```python
print(kmeans.predict(x))
```

```python
# cluster_centers_  ->   
mglearn.discrete_scatter(x[:, 0], x[:, 1], kmeans.labels_,
                        markers='o') #  
mglearn.discrete_scatter(kmeans.cluster_centers_[:, 0],
                         kmeans.cluster_centers_[:, 1],
                         [0,1,2], markers='^', markeredgewidth=5) #  
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))

#  2
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
mglearn.discrete_scatter(x[:,0], x[:,1], kmeans.labels_, ax=axes[0])

#  5
kmeans = KMeans(n_clusters=5)
kmeans.fit(x)
mglearn.discrete_scatter(x[:,0], x[:,1], kmeans.labels_, ax=axes[1])
```
