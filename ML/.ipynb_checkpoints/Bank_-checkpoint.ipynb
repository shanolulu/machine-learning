```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('bank-additional-full.csv', delimiter=';')
train = train.drop(['duration'], axis=1)
# header = ['age', 'campaign', 'pdays', 'previous', 'emp.var.rate',
#           'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
# data.hist(column=header, figsize=(10, 10))
```

```python
test = pd.read_csv('bank-additional.csv', delimiter=';')
test = test.drop(['duration'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
train['job'] = label.fit_transform(train['job'])
train['marital'] = label.fit_transform(train['marital'])
train['education'] = label.fit_transform(train['education'])
train['default'] = label.fit_transform(train['default'])
train['housing'] = label.fit_transform(train['housing'])
train['loan'] = label.fit_transform(train['loan'])
train['contact'] = label.fit_transform(train['contact'])
train['month'] = label.fit_transform(train['month'])
train['day_of_week'] = label.fit_transform(train['day_of_week'])
train['poutcome'] = label.fit_transform(train['poutcome'])

test['job'] = label.fit_transform(test['job'])
test['marital'] = label.fit_transform(test['marital'])
test['education'] = label.fit_transform(test['education'])
test['default'] = label.fit_transform(test['default'])
test['housing'] = label.fit_transform(test['housing'])
test['loan'] = label.fit_transform(test['loan'])
test['contact'] = label.fit_transform(test['contact'])
test['month'] = label.fit_transform(test['month'])
test['day_of_week'] = label.fit_transform(test['day_of_week'])
test['poutcome'] = label.fit_transform(test['poutcome'])
```

```python
x_train = train.iloc[:, :19]
y_train = train.iloc[:, 19:20]

x_test = test.iloc[:, :19]
y_test = test.iloc[:, 19:20]
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(C=10)
svm.fit(x_train, y_train) # predict -> accuracy_score
print('테스트 세트 정확도: {:.4f}'.format(svm.score(x_test, y_test)))
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred = svm.predict(x_test)
pred
```

```python
print(type(y_test))
print(type(pred))
```

```python
print('테스트 세트 정확도: {:.4f}'.format(accuracy_score(y_test, pred)))
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier

lgb = LGBMClassifier()
lgb_param_grid = {
    'n_estimators' :[100, 200, 300],
    'max_depth': [5, 10, 15, 20],
    'min_child_samples': [10, 50, 100],
    'subsample': [0.8, 1, 1.2],
    'max_bin':[5, 10, 30, 50, 100, 300],
    'boost': ['dart']
}
lgb_grid = GridSearchCV(lgb, param_grid=lgb_param_grid)
```

```python
lgb_grid.fit(x_train, y_train)
```

```python
print('테스트 세트 정확도: {:.4f}'.format(lgb_grid.score(x_test, y_test)))
```
