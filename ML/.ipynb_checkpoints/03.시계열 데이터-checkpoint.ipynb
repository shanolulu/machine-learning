```python
import pandas as pd
df = pd.read_csv('example_retail_sales.csv', encoding='cp949')

# 시계열 데이터: 일정 시간 간격으로 배치된 데이터들의 수열
from fbprophet import Prophet
m = Prophet()
m.fit(df)
```

```python
df.tail()
```

```python
df.describe()
```

```python
future = m.make_future_dataframe(periods=365) # 데이터가 365개 더 늘어남
```

```python
future.tail()
```

```python
forecast = m.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail() # yhat: predict
```

```python
fig1 = m.plot(forecast)
```

```python
fit2 = m.plot_components(forecast)
```
