```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import datasets
import pandas as pd
```

```
cancer = datasets.load_breast_cancer()
cancer
```

```
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df
```

```
# 학습데이터와 검증데이터 분리
x = cancer['data']
y = cancer['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=cancer['target'], random_state=0)
```

```
# 기본 값이 max_depth=3, learning_rate=0.1
model = GradientBoostingClassifier(random_state=0)
# 학습
model.fit(x_train, y_train)
```

```
score_train = model.score(x_train, y_train)
print('score_train=', score_train)
score_test = model.score(x_test, y_test)
print('score_test=', score_test)
```

```
# 기본 값이 max_depth=3, learning_rate=0.1
model = GradientBoostingClassifier(random_state=0, max_depth=1)
# 학습
model.fit(x_train, y_train)
score_train = model.score(x_train, y_train)
print('max_depth=1', score_train)
score_test = model.score(x_test, y_test)
print('max_depth=1', score_test)
```

```
# 기본 값이 max_depth=3, learning_rate=0.1
model = GradientBoostingClassifier(random_state=0, max_depth=5, learning_rate=0.05)
# 학습
model.fit(x_train, y_train)
score_train = model.score(x_train, y_train)
print('max_depth=5, learning_rate=0.05', score_train)
score_test = model.score(x_test, y_test)
print('max_depth=5, learning_rate=0.05', score_test)
```

```
# 기본 값이 max_depth=3, learning_rate=0.1
model = GradientBoostingClassifier(random_state=0, max_depth=3, learning_rate=0.01)
# 학습
model.fit(x_train, y_train)
score_train = model.score(x_train, y_train)
print('max_depth=3, learning_rate=0.01', score_train)
score_test = model.score(x_test, y_test)
print('max_depth=3, learning_rate=0.01', score_test)
```
