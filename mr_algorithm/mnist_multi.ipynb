```
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```
train_df = pd.read_csv('/content/drive/My Drive/SKinfosec/mnist/mnist_train.csv')
test_df =pd.read_csv('/content/drive/My Drive/SKinfosec/mnist/mnist_test.csv')
```

```
train_data = train_df.values
# 8;2 -> 학습데이터 : 검증데이터
x_train, x_test, y_train, y_test = train_test_split(train_data[:, 1:], train_data[:, 0], test_size=0.2, random_state=0)
```

# RandomForestClassifier

```
# train_df의 데이터로 predict
model1 = RandomForestClassifier(n_estimators=2000)
rdc = model1.fit(x_train, y_train)
output1 = rdc.predict(x_test)
```

```
print('RandomForest with n_estimators=2000')
print(accuracy_score(y_test, output1))
```

```
# 1) test_df의 데이터로 predict
train_df_x = train_data[:, 1:]
train_df_y = train_data[:, 0]
test_df_x = test_data[:, 1:]
test_df_y = test_data[:, 0]

model1 = RandomForestClassifier(n_estimators=2000)
rdc = model1.fit(train_df_x, train_df_y)
output1 = rdc.predict(test_df_x)
print('RandomForest with n_esimators=2000')
print(accuracy_score(test_df_y, output1))
```

```
# 2) 학습데이터는 train_df, 검증데이터는 test_df로 작업
model1 = RandomForestClassifier(n_estimators=500)
rdc = model1.fit(x_train, y_test)
output1 = rdc.predict(x_test)
```

# GradientBoostingClassifier

```
model2 = GradientBoostingClassifier(n_estimators=10, max_depth=1, learning_rate=0.3, random_state=0)
gdc = model2.fit(x_train, y_train)
output2 = gdc.predict(x_test)
```

```
print('GradientBoostingClassifier with n_estimators=10, max_depth=1, learning_rate=0.3')
print(accuracy_score(y_test, output2))
```

# MLPClassifier

```
model3 = MLPClassifier(solver='sgd', hidden_layer_sizes=(100,), random_state=1)
mpc = model3.fit(train_df_x, train_df_y)
```

```
output3 = mpc.predict(test_df_x)
```

```
print('MPClassifier with ')
print(accuracy_score(test_df_y, output3))
```

```
model3 = MLPClassifier(solver='adam', hidden_layer_sizes=(100,), random_state=1)
mpc = model3.fit(train_df_x, train_df_y)
output3 = mpc.predict(test_df_x)
```

```
print('MPClassifier with ')
print(accuracy_score(test_df_y, output3))
```

```
import csv
final_output = model1
predictions_file = open('/content/drive/My Drive/SKinfosec/forest_output.csv', 'w')
open_file_object = csv.writer(predictions_file)
ids = range(output1.__len__())
ids = [x+1 for x in ids]
open_file_object.writerow(['ImageId', 'Label'])
open_file_object.writerow(zip(ids,final_output))
predictions_file.close()
print('Saved forest_output file')
```
